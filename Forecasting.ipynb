```python
df_new = df[['ds','y']] #create a new dataframe of only these two columns
df_new
```

```python
df_new['ds']=pd.to_datetime(df_new['ds'])
df_new.info()
```

```python
df_new['y'] = df_new['y'].astype(int) #object to int datatype conversion
df_new.info()
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize = (20,10)) #increase fiigure size 
plt.scatter(df_new['ds'],df_new['y']) # values from the dataset 
plt.show()
```

```python
import numpy as np
df_new['y']=np.where(df_new['y']>2700,np.nan,df_new['y'])
df_new
```

```python
df_new.isnull().sum()
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize = (20,10)) #increase fiigure size 
plt.scatter(df_new['ds'],df_new['y']) # values from the dataset 
plt.show()
```

```python
import fbprophet
op=fbprophet.Prophet()
op=fbprophet.Prophet(changepoint_prior_scale=1)
op.fit(df_new)
```

```python
forecast=op.make_future_dataframe(periods=365,freq='D')
```

```python
forecast=op.predict(forecast)
```

```python
forecast
```

```python
plt.figure(figsize=(25,20))
op.plot(forecast,xlabel='Price',ylabel='Date')
```
